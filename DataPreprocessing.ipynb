# ICRISAT Data Set Preprocessing

```python
import pandas as pd

# Read CSV file of icrisat into a DataFrame
df_icrisat = pd.read_csv('https://raw.githubusercontent.com/Cropdata5320/CropData_Visualizations/main/ICRISAT-District%20Level%20Data.csv')
print(df_icrisat.columns)
df_icrisat.head()
print(df_icrisat.shape)
```

```python
df_icrisat.info()
```

```python
# Check for null values in each column
print("Null values in each column:")
df_icrisat.isnull().sum()
```

```python
# Remove rows with null values
df_icrisat = df_icrisat.dropna()
df_icrisat.isnull().sum()
```

```python
# Write DataFrame back to a CSV file
df_icrisat.to_csv('ICRISAT_cleaned.csv', index=False)
```

# Food_prices Data set Preprocessing

```python
df_food_prices = pd.read_csv('https://raw.githubusercontent.com/Cropdata5320/CropData_Visualizations/main/wfp_food_prices_ind.csv', dtype={'date': str})
print(df_food_prices.columns)
df_food_prices.head()
```

```python
df_food_prices.info()
print(df_food_prices.shape)
```

```python
# Drop the first row
df_food_prices = df_food_prices.drop(index=0)
print(df_food_prices.columns)
df_food_prices.head()
```

```python
# Rename the column 'admin1' to 'state' and 'admin2' to 'city'
df_food_prices = df_food_prices.rename(columns={'admin1': 'state', 'admin2': 'city'})

# Remove 'market' & 'pricetype' columns
df_food_prices = df_food_prices.drop(columns=['market', 'pricetype']).reset_index(drop=True)

df_food_prices.columns
```

```python
# Convert the date column to datetime format
df_food_prices['date'] = pd.to_datetime(df_food_prices['date'], format='%Y-%m-%d', errors='coerce')

# Extract the year from the datetime column and add it as a new column beside the date column
df_food_prices.insert(df_food_prices.columns.get_loc('date') + 1, 'Year', df_food_prices['date'].dt.year)

df_food_prices.head()
```

```python
# Check for null values in each column
print("Null values in each column:")
df_food_prices.isnull().sum()
```

```python
# Remove rows with null values
df_food_prices = df_food_prices.dropna()
df_food_prices.isnull().sum()
```

```python
# Save the DataFrame to a new CSV file
df_food_prices.to_csv('Food_Prices_cleaned.csv', index=False)
df_food_prices.info()
```
