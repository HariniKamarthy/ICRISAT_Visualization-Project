```python
import pandas as pd

# Load the datasets
combined_data = pd.read_csv('https://raw.githubusercontent.com/Cropdata5320/CropData_Visualizations/main/CropYield_Weather_Prices_combined.csv')

# Since we need to group by 'State Name' and 'Year' and calculate the average of certain columns,
# we perform the following groupby operation
grouped = combined_data.groupby(['State Name', 'Year']).agg({
    'yield': 'mean',
    'Avg_Annual_Temperature': 'mean',
    'Avg_Annual_Rainfall': 'mean',
    'average_price': 'mean'
}).reset_index()

grouped
```

```python
# Save the DataFrame to a new CSV file
grouped.to_csv('Task6_ProcessedData.csv', index=False)
```
