```python
import pandas as pd

# Load the datasets
icrisat_data = pd.read_csv('https://raw.githubusercontent.com/Cropdata5320/CropData_Visualizations/main/ICRISAT_cleaned.csv')

extract_df = icrisat_data[['Year', 'State Name', 'RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)', 'RICE YIELD (Kg per ha)']]

extract_df.tail()
```

```python
# Group the data by 'Year' and 'State Name' and calculate the mean for each group
grouped_t7_df = extract_df.groupby(['Year', 'State Name']).agg({
    'RICE AREA (1000 ha)': 'mean',
    'RICE PRODUCTION (1000 tons)': 'mean',
    'RICE YIELD (Kg per ha)': 'mean'
}).reset_index()

grouped_t7_df.tail()
```

```python
# Group the data by 'Year' and 'State Name' and calculate the mean for each group
grouped2_t7_df = grouped_t7_df.groupby('State Name').agg({
    'RICE AREA (1000 ha)': 'mean',
    'RICE PRODUCTION (1000 tons)': 'mean',
    'RICE YIELD (Kg per ha)': 'mean'
}).reset_index()

grouped2_t7_df
```

```python
# Save the DataFrame to a new CSV file
grouped2_t7_df.to_csv('Task7_ProcessedData.csv', index=False)
```
